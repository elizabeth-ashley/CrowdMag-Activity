```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import os, sys
pd.options.display.float_format = '{:.2f}'.format

directory = './data/'

for file in os.listdir(directory):
    df = pd.read_csv(directory + file)
    northing = df['Latitude (deg)']*111300
    easting = df['Longitude (deg)']*85300

    df.insert(loc=6,
              column='Northing (m)',
              value=northing)

    df.insert(loc=7,
              column='Easting (m)',
              value=easting)

    df2 = df.sort_values(by=['Northing (m)'], ascending=True)
    df3 = df.sort_values(by=['Easting (m)'], ascending=True)


    fig = make_subplots(rows=2, cols=1)

    fig.append_trace(go.Scatter(x = df2['Northing (m)'], y = df2[' Total Field (nT)'], name="Northing (m)")
                     , row=1, col=1)             
    fig.append_trace(go.Scatter(x = df3['Easting (m)'], y = df3[' Total Field (nT)'], name="Easting (m)")
                     , row=2, col=1)

    fig.update_layout(height=600, width=600, title_text='Total Field Anomaly (nT)'+ ' - ' + file, )
    fig.update_layout(xaxis=dict(tickformat ='d'), xaxis2=dict(tickformat ='d'), yaxis=dict(tickformat ='d'), yaxis2=dict(tickformat ='d'))
    fig.write_image("plot_"+file+".pdf")
    fig.show()
        
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import os, sys
pd.options.display.float_format = '{:.2f}'.format

directory = './data/'
filenames = []
filelist = []
for file in os.listdir(directory):
    filenames = directory + file 
    filelist.append(filenames)
# print(filelist)
    
    
#  dict for the dataframes and their names
dfs = {"Albertoni" : [], "Carter": [], "Grossaint" : [], "Kissner" : [], "Qasemi" : [], "Riccio" : [], "Saltus" : []}  
i = 0 
for key in dfs:
    df = pd.read_csv(filelist[i])
                    
    northing = df['Latitude (deg)']*111300
    easting = df['Longitude (deg)']*85300

    df.insert(loc=6,
              column='Northing (m)',
              value=northing)

    df.insert(loc=7,
              column='Easting (m)',
              value=easting)
    
    dfs[key] =  df.sort_values(by=['Northing (m)'], ascending=True)
    i = i + 1
    

# plot the data
fig = go.Figure()

for d in dfs:
    fig = fig.add_trace(go.Scatter(x = dfs[d]['Northing (m)'],
                                   y = dfs[d][' Total Field (nT)'], 
                                   name = d))
    fig.update_layout(title_text='Magnetic N-S Profile'+ ' - all.csv',
                     xaxis_title="Northing (m)",
                     yaxis_title="Total Field Anomaly (nT)",
                     legend_title="Data Collector")
    fig.update_layout(xaxis=dict(tickformat ='d'), xaxis2=dict(tickformat ='d'), yaxis=dict(tickformat ='d'), yaxis2=dict(tickformat ='d'))
fig.write_image("plot_Magnetic N-S Profile_all.png")
fig.show()

   
```
